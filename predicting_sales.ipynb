```python
x=X.drop(columns='sales_p_month')
y=X.sales_p_month
```

```python
shopname = pd.get_dummies(x['shop_name'],drop_first=True)

x = pd.concat([x.drop('shop_name',axis=1),shopname],axis=1)
x
```

```python
categ = pd.get_dummies(x['item_category_name'],drop_first=True)

x = pd.concat([x.drop('item_category_name',axis=1),categ],axis=1)
x
```

```python
x.drop(columns='ID',inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

numscale = make_column_transformer((MinMaxScaler(),['item_price']))
```

# Modeling


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()



# input layer
model.add(Dense(145,  activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(39, activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(19, activation='relu'))
model.add(Dropout(0.2))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(loss='mse', 
              optimizer='adam', 
              metrics=['mae'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=4,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=True
)
```

```python
pipe = make_pipeline(numscale,model)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test,y_pred)
```
